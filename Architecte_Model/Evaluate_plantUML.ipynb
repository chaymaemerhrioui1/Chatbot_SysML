2. Render and Save Diagrams

```python
import os
import subprocess
from IPython.display import Image, display

# Create a directory to save the diagrams
output_dir = '/content/plantuml_diagrams'
os.makedirs(output_dir, exist_ok=True)


# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    plantuml_code = row['uml_code'].replace('""', '"')

    # Define the output file path
    output_file = os.path.join(output_dir, f'diagram_{index + 1}.png')

    try:
        # Write the PlantUML code to a temporary file
        with open('temp.puml', 'w') as f:
            f.write(plantuml_code)

        # Run PlantUML to generate the diagram
        result = subprocess.run(['java', '-jar', 'plantuml.jar', 'temp.puml', '-o', output_dir], check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)

        # If successful, move the file to the desired location
        os.rename(os.path.join(output_dir, 'temp.png'), output_file)

        # Update the status
        df.at[index, 'render_status'] = 'Rendered Successfully'

        # Display the rendered diagram
        display(Image(filename=output_file))

    except subprocess.CalledProcessError as e:
        # If there is an error, capture the error message
        error_message = e.stderr
        df.at[index, 'render_status'] = f'Error: {error_message}'

        # Optionally, you can print the error message
        print(f"Error rendering diagram {index + 1}: {error_message}")

    except Exception as e:
        # Handle other possible exceptions
        df.at[index, 'render_status'] = f'Unexpected Error: {str(e)}'
        print(f"Unexpected error rendering diagram {index + 1}: {str(e)}")
```

3. Evaluate semantic

```python
import requests
import pandas as pd
import json
import time
import random
from google.colab import userdata

# Replace with your actual Groq API key
GROQ_API_KEY = userdata.get('GROQ_API_KEY')

# Groq API endpoint
GROQ_API_URL =  "https://api.groq.com/openai/v1/chat/completions"

# Headers for the API request
headers = {
    'Authorization': f'Bearer {GROQ_API_KEY}',
    'Content-Type': 'application/json'
}
```

```python
def evaluate_semantic_groq(description, plantuml_code):
    # Define the prompt
    prompt = (
        f"Description:\n{description}\n\n"
        f"PlantUML Code:\n{plantuml_code}\n\n"
        f"Question: Does the PlantUML code accurately represent the - Participants - Messages exchanged and their precise chronology- Conditions, loops, and alternative pathsdescription? "
        f"Answer with 'Yes' or 'No' don't be harsh."
    )

    # Define the payload for the API request
    payload = {
        "model": "llama-3.3-70b-versatile",
        "messages": [{"role": "user", "content": prompt}],
        "temperature": 0.2,
        "max_tokens": 2048
    }

    try:
        # Make a POST request to the Groq API
        response = requests.post(GROQ_API_URL, headers=headers, data=json.dumps(payload))
        response.raise_for_status()
        # Raise an exception for HTTP errors
        response.raise_for_status()

        # Parse the JSON response
        data = response.json()

        # Extract the generated text
        generated_text = data["choices"][0]["message"]["content"].strip().lower()
        print("--------------------------------",generated_text,"--------------------------------")

        # Determine the result
        is_correct = generated_text == 'yes.'

        return is_correct

    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred: {http_err}")
    except Exception as err:
        print(f"An error occurred: {err}")
```

```python
# Set the number of samples to evaluate
num_samples = 4

samples = df.sample(n=num_samples, random_state=42)

# Iterate through the samples and evaluate
for index, row in samples.iterrows():
    description = row['description']
    plantuml_code = row['uml_code']
    is_correct = evaluate_semantic_groq(description, plantuml_code)
    df.at[index, 'semantic_status'] = 'Correct' if is_correct else 'Incorrect'

    print(f"Sample {index + 1}:")
    print(f"Description: {description}")
    print(f"PlantUML Code:\n{plantuml_code}\n")
    print(f"Semantic Correctness: { is_correct}\n")
```
